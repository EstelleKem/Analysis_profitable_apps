2. Since the compagny it only concern with *`english`* and`free` apps, we will write codes not only to retrieve `english apps` but also those who are `free`.

    a. According to the ASCII (American Standard Code for Information Interchange) system, the numbers corresponding to the characters commonly use in an English text are all in the range 0 to 127, . Based on this number range, the code below was written to identify wheter an application is directed to an *`English-speaking`* audience or not. 
    > According to the result obtain from the code written below, some Applications, which seem to be *`english`* applications have characters with an ASCII range greater than 127. As an example the application named `U Launcher Lite – FREE Live Cool Themes, Hide Apps ` in the `Androids_apps` dataset and `The Sims™ FreePlay 8482` in the `iOS_apps` dataset have respectively characters with an ASCII range of 8211 and 8482. This implies that, an additional condition should be set in order to obtain a better segregation between *`english`* and *`non-english Apps`*.
    
    > To Avoid the wrong labeling of *`english`* and *`non-english Apps`*, a new condition - in addition to the ASCII range- was defined. This condition basically defines *`non-english Apps`* as those apps whose names have more than 3 characters with a range greater than 127. 

**Androids Dataset**

```python
# identify the non english Apps based on the AscII range for english text.
Androids_non_english_apps = []
for row in Androids_apps_clean_1:
    count_range_letter_tot = 0
    AppsName = row [0]
    for letter in AppsName:
        range_letter = ord(letter)
        if range_letter > 127:
            count_range_letter_tot += 1 

#  setting a new condition for a better labeling
    if count_range_letter_tot > 3:
        Androids_non_english_apps.append(row)
    
        

print(Androids_non_english_apps[:2])
print(len(Androids_non_english_apps))
print(len(Androids_apps_clean_1))
              
```

**iOS Dataset**

In the following a frequency datasets of the application *`genre`* ,*`category`*, and *`prime genre`* is build, so as to draw first conclusion on the type of applications mostly use on both market.

**Androids Dataset**

`Variable genre` and `category`

 * From the result obtain below, the five mostly used genre of the Android dataset are `Tools`, `Entertainment`, `Education`, `Business` and `Productivity` with frequencies of  `749`,`538`,`474`,`407`,`345` respectively.

* The five most used applications categories are `Family`, `Game`, `Tools`, `Business` and `Lifestyle`, representing respectively`18,9%`, `9,7%`, `8,5%`, `4,6%` and `4%` of the total.


```python
def column_frequency (dataset,selected_row, total_apps):
    Apps_dic = {}
    Apps_list = []
    Apps_frequency_tuple = []

# determine the frequency of the selected variable of the dataset
    for row in dataset:
        Apps_row = row[selected_row]
        Apps_list.append(Apps_row)
    for element in Apps_list:
        if element in Apps_dic:
            Apps_dic[element] += 1
        else:
            Apps_dic[element] = 1
        
# consigne the frequency in a data type 'tuple' and the value at index '0' to make the sorting easier
    for keys in Apps_dic:
        value = Apps_dic[keys]
        value_percentage = (Apps_dic[keys]/total_apps)* 100
        Apps_frequency_tuple.append((value,round(value_percentage,2), keys))
    
    return sorted(Apps_frequency_tuple, reverse = True)

Androids_frequency_genre = column_frequency(Androids_english_free_apps, 9,8864)

Androids_frequency_genre[:10]
```

In the following, applications with the highest number of users will be determined. 
* To reach the above mentioned aim, we could basically calculate the average number of `installs` for each app genre. If this information is availabe for the `Androids` dataset, it is unfortunately missing the `iOS` dataset.

* So instead of using the average number of `installs` for the `iOS` dataset , we'll take the total number of user ratings (`rating_count_tot`).

**Androids Dataset**

```python
# function to build a dictionary with 'genre' as keys and 'ratings' or 'installs' as values
def genre_rating_or_installs(dataset, genre_column,):
    dic = {}
    for row in dataset:
        Apps_genre = row[genre_column]
        rating_or_installs = row[5]
        if Apps_genre in dic: 
            dic[Apps_genre].append(rating_or_installs) 
        else:
            dic[Apps_genre] = [rating_or_installs]
    return dic

Androids_installs_genre_dic = genre_rating_or_installs(Androids_english_free_apps, 9)


# check if the code produces the expected result
for key in  Androids_installs_genre_dic:
        if key == 'Tools':
            print(len(Androids_installs_genre_dic[key]))
```

* Looking at the number of `Installs` of the `Androids` dataset, we realise that most of the values are open-ended (100+, 1,000+, 5,000+, etc.), lacking therefore of precision. It is obviously not clear, if an application with 1,000+ has exactly 1,000 installations or more(1,500 or 2,500). Yet, since our goal is to find out which application genres are the most popular(have the highest number of users), we will consider that an application with `n+` installs has `n` installs

> To perform computations, however, we'll need to convert each install number from string to float. This means we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error.

```python
#  function to obtain the average 'rating' or 'installs' of each application genre
def avg_install_or_ratings(dataset):
    Avg_genre_tuple = []
    for key in dataset:
        total = 0
        len_value = 0
# remove the comma and the plus sign from the number of `installs`    
        for element in dataset[key]:
            element = element.replace('+','')
            element = element.replace(',', '')
            total += float(element)
            len_value += 1
# calculation the average number of `installs` for each genre
        Avg = total/len_value
        Avg_genre_tuple.append((round(Avg,2),key))
    
#  print the sorted average values in descending order 
    return sorted(Avg_genre_tuple, reverse = True)[:10]
            

Androids_avg_installs_genre = avg_install_or_ratings(Androids_installs_genre_dic)
Androids_avg_installs_genre
```

the result obtained above, clearly shows that, the `communication` genre is the most populare genre in the Androids dataset, follow by the `Adventure;Action & Adventure'` and the `Video Players & Editors` genre.

> the popularity of the `Communication` genre is highly influenced by a few apps that have over one billion installs (Messenger – Text and Video Chat for Free, WhatsApp Messenger, Viber Messenger,...).

```python
def apps_most_popular_genre(dataset, AppsName_row, Apps_ratings_row, Apps_genre_row, Apps_genre):
    for row in dataset:
        AppsName = row[AppsName_row] 
        AppsPrimeGenre = row[Apps_genre_row]
        AppsRating = row[Apps_ratings_row]
    
        if  AppsPrimeGenre == Apps_genre:
            print (AppsName, ':', AppsRating)
    print("\n")
       
Navigation_apps =  apps_most_popular_genre(iOS_english_free_apps, 1, 5, 11, 'Navigation') 
Reference_apps = apps_most_popular_genre(iOS_english_free_apps, 1, 5, 11, 'Reference')
SocialNetworking_apps = apps_most_popular_genre(iOS_english_free_apps, 1, 5, 11, 'Social Networking')
```
